```python
# import the load_boston() function from sklearn.datasets
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()

print(boston.data)
print(boston.data.shape)

bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
bos['PRICE'] = boston.target

print(bos.head())
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf
```

```python
feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
        'TAX', 'PTRATIO', 'B', 'LSTAT']
X = bos[feature_cols]
y = bos["PRICE"]
```

```python
lm = LinearRegression()
lm.fit(X, y)
new_X = X.values.reshape(-1, 1)
new_y = y.values.reshape(-1, 1)

# new_X = X.values.reshape((6578,506))
# new_y = y.values.reshape((6578,506))

# reg = LinearRegression().fit(new_X, new_y)
reg = LinearRegression().fit(new_X[:506], new_y)
print(len(new_X[:506]))
print(len(new_X))
print(len(new_y))
```

```python

print(lm.intercept_)
print(lm.coef_)
#Implementing a linear regression with mutliple features in relation to the MEDV or "PRICE" feature
```

```python
import statsmodels.formula.api as smf
```

```python
lm = smf.ols(formula='PRICE ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data=bos).fit()
```

```python
# print the coefficients
lm.params
```

```python
from sklearn.metrics import r2_score

print(reg.score(new_X[:506].reshape(-1, 1), new_y.reshape(-1, 1)))
print(r2_score(new_y.reshape(-1, 1), reg.predict(new_X[:506].reshape(-1, 1))))
```
