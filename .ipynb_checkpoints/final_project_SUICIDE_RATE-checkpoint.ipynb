```python
total_suicides = 0
for number in df['suicides_no']:
    total_suicides += number
print(total_suicides)    
```

### the total number of suicides for all countries from 1987 to 2016 is 6,748,420.

```python
num_suicides_per_country = []

current_country = "Albania"
count = 0
for i, country in enumerate(df['country']):
    if current_country == country:
        count += df["suicides_no"][i]
    else:
        num_suicides_per_country.append((current_country,count))
        count = df["suicides_no"][i]
        current_country = country
        
print(num_suicides_per_country)

count = 0
for country, suicide_sum in num_suicides_per_country:
    count+= suicide_sum

print(count)
        
        
```

### total number of suicides found above (in cell 61) == 6,748,420

### total number of suicides found above (in cell 68) == 6,713,617

### similar numbers, but slightly different. where is the discrepency coming from?

### 6,748,420 != 6,713,617

```python
max_suicides = [None, 0]
for country, suicide_sum in num_suicides_per_country:
    max_suicides[1] = max(suicide_sum, max_suicides[1])
    if max_suicides[1] == suicide_sum:
        max_suicides[0] = country

print(max_suicides)
```

### The "Russian Federation" has the highest rate of suicides from 1987 to 2016, with 1,209,742 reported suicides.

```python
suicide_lookup_dictionary = {}
array_for_heap = []

for country, suicide_sum in num_suicides_per_country:
    suicide_lookup_dictionary[suicide_sum] = country
    array_for_heap.append(suicide_sum)
    
heapq.heapify(array_for_heap)
five_most_suicidal_countries = heapq.nlargest(5,array_for_heap)

for i, suicide_number in enumerate(five_most_suicidal_countries):
    print(str(suicide_lookup_dictionary[suicide_number]) + " had " + str(suicide_number))
```

## the five most suicidal countries

### Russian Federation had 1,209,742

### United States had 1,034,013

### Japan had 806,902

### France had 329,127

### Ukraine had 319,950

```python
suicide_lookup_dictionary = {}
array_for_heap = []

for country, suicide_sum in num_suicides_per_country:
    suicide_sum = suicide_sum * -1
    suicide_lookup_dictionary[suicide_sum] = country
    array_for_heap.append(suicide_sum)
    
heapq.heapify(array_for_heap)
five_least_suicidal_countries = heapq.nlargest(5,array_for_heap)

for i, suicide_number in enumerate(five_least_suicidal_countries):
    print(str(suicide_lookup_dictionary[suicide_number]) + " had " + str(-1*suicide_number))
```

```python
curr_country_year = df['country-year'][0]
country_year_suicides_100k_average_dictionary = {}
percentage_sum = 0
len_count = 0
for i, country_year in enumerate(df['country-year']):
    if country_year != curr_country_year:
        country_year_suicides_100k_average_dictionary[curr_country_year] = float(percentage_sum/len_count)
        percentage_sum = 0
        len_count = 0
        curr_country_year = country_year
    percentage_sum += df['suicides/100k pop'][i]
    len_count += 1

print(country_year_suicides_100k_average_dictionary).entries()
        
```

## Looking at the suicides per 100k population percentage and averaging the percentages by country-year as this seems to be the best metric for comparison between countries.

## The number of suicides should be relative to a country's population size.

```python
curr_country_year = df['country-year'][0]
country_year_gdp_per_capita_dictionary = {}
gdp_sum_by_country_year = 0
len_count = 0
for i, country_year in enumerate(df['country-year']):
    if country_year != curr_country_year:
        country_year_gdp_per_capita_dictionary[curr_country_year] = float(gdp_sum_by_country_year/len_count)
        gdp_sum_by_country_year = 0
        len_count = 0
        curr_country_year = country_year
    gdp_sum_by_country_year += df['gdp_per_capita ($)'][i]
    len_count += 1

print(country_year_gdp_per_capita_dictionary).entries()
```

## "Gross domestic product (at purchasing power parity) per capita, i.e., the purchasing power parity (PPP) value of all final goods and services produced within a country in a given year, divided by the average (or mid-year) population for the same year."

https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita

## The above model is wrong due to lack of splitting the data ALSO!! country_year_gdp_per_capita_dictionary.values() and country_year_suicides_100k_average_dictionary.values() returns the values in a random order so the data between the two ISN'T being matched up correctly despite having the same number of entries and the same keys.

```python
X = []
y = []

for key in country_year_gdp_per_capita_dictionary.keys():
    # values are being put into respective arrays in same order with same index.
    X.append(country_year_gdp_per_capita_dictionary[key])
    y.append(country_year_suicides_100k_average_dictionary[key])

X = np.array(X)
y = np.array(y)


plt.scatter(X, y)
plt.xlabel('gdp_per_capita')
plt.ylabel('Percentage of Suicides out of 100k')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
lr_reg = LinearRegression()

lr_reg.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))

print(lr_reg.coef_) #m

print(lr_reg.intercept_) #b

# m = -0.000001.77630687
# b = 12.84006532

plt.plot(X_train, y_train, 'ro')
plt.axis([0, 120000, 0, 60])
plt.show()
```

```python
# https://stackoverflow.com/questions/14000595/graphing-an-equation-with-matplotlib
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = eval(formula)
    plt.plot(x, y)  
    plt.show()

graph('-1.77630687e-06*x+12.84006532', X_test)
```

```python
print(lr_reg.predict(X_test.reshape(-1, 1)))

print(lr_reg.score(X_test.reshape(-1, 1), y_test.reshape(-1, 1)))
print(r2_score(y_test, lr_reg.predict(X_test.reshape(-1, 1))))
```
